```python
# Sample quiz questions and answers
quiz_questions = [
    {
        'question': 'How long does a typical menstrual cycle last?',
        'options': ['21-35 days', '15-20 days', '40-50 days'],
        'correct_answer': '21-35 days'
    },
    {
        'question': 'What is a common symptom of PMS?',
        'options': ['Nausea', 'Euphoria', 'Fatigue'],
        'correct_answer': 'Fatigue'
    },
]

# Function to conduct the quiz
def conduct_quiz(quiz):
    score = 0
    for q in quiz:
        print(q['question'])
        for i, option in enumerate(q['options']):
            print(f"{i + 1}. {option}")
        answer = input("Enter the option number: ")
        if q['options'][int(answer) - 1] == q['correct_answer']:
            score += 1
            print("Correct!")
        else:
            print("Incorrect.")
        print()
    print(f"Your score: {score}/{len(quiz)}")

# Conduct the quiz
conduct_quiz(quiz_questions)
```
